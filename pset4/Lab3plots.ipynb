```python
#making some plots for my students for their lab 3 

import numpy as np 
import matplotlib.pyplot as plt 
import astropy.units as u 
```

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']

a = [0.39, 0.72, 1, 1.52, 5.20, 9.54, 19.19, 90.06]
e = [0.206, 0.007, 0.017, 0.093, 0.048, 0.056, 0.046, 0.010]
```

```python
%config InlineBackend.figure_format = "retina"

plt.rc('font', family='serif')
plt.rcParams['text.usetex'] = False
fs = 18

# update various fontsizes to match
params = {'figure.figsize': (12, 8),
          'legend.fontsize': fs,
          'axes.labelsize': fs,
          'xtick.labelsize': 0.9 * fs,
          'ytick.labelsize': 0.9 * fs,
          'axes.linewidth': 1.1,
          'xtick.major.size': 7,
          'xtick.minor.size': 4,
          'ytick.major.size': 7,
          'ytick.minor.size': 4}
plt.rcParams.update(params)
```

```python
#Question 2 (Orbital Eccentricity vs Semi-major axis)
plt.scatter(a,e)
plt.plot(a[2],e[2], 'o', markersize = 15, color = 'g', label = planets[2])
plt.plot(a[4],e[4], 'o', markersize = 30, color = 'orange', label = planets[4])
plt.title('Solar System Data for Question 2')
plt.ylabel('Orbital Eccentricity')
plt.xlabel('Semi-Major Axis')
plt.grid()
plt.legend()
plt.show()
```

# The vales are very very low and they (for the most part decrease as you increase semi-major axis). There is a small increase from Earth to Jupiter, but on the whole all the values are low when compared to the Exoplanets which appear to have to coherent pattern or trend

```python
plt.semilogy(a,e, 'o')
plt.plot(a[2],e[2], 'o', markersize = 15, color = 'g', label = planets[2])
plt.plot(a[4],e[4], 'o', markersize = 30, color = 'orange', label = planets[4])
plt.title('Solar System Data for Question 2')
plt.ylabel('Orbital Eccentricity')
plt.xlabel('Semi-Major Axis')
plt.grid()
plt.legend()
plt.show()
```

```python
#Question 3 (Planetary Mass vs Orbital Eccentricity)

mass = [0.003146, 1, 0.299, 0.045,0.053]

plt.plot(a[2],mass[0], 'o', markersize = 15, color = 'g', label = planets[2])
plt.plot(a[4],mass[1], 'o', markersize = 30, color = 'orange', label = planets[4])
plt.plot(a[5],mass[2], 'o', markersize = 20, color = 'pink', label = planets[5])
plt.plot(a[6],mass[3], 'o', markersize = 20, color = 'blue', label = planets[6])
plt.plot(a[7],mass[4], 'o', markersize = 20, color = 'purple', label = planets[7])
plt.title('Solar System Data for Question 3')
plt.ylabel('Planetary Mass [Mjup]')
plt.xlabel('Semi-Major Axis')
plt.grid()
plt.legend()
plt.show()
```

# ALL the values are VERY low (below 1). When compared to the Exoplanets (which have no coherent pattern) the only relationship the Solar system has is that as you increase semi major axis they have about the same/"constant" planetary mass

```python
plt.semilogy(a[2],mass[0], 'o', markersize = 15, color = 'g', label = planets[2])
plt.semilogy(a[4],mass[1], 'o', markersize = 30, color = 'orange', label = planets[4])
plt.semilogy(a[5],mass[2], 'o', markersize = 20, color = 'pink', label = planets[5])
plt.semilogy(a[6],mass[3], 'o', markersize = 20, color = 'blue', label = planets[6])
plt.semilogy(a[7],mass[4], 'o', markersize = 20, color = 'purple', label = planets[7])
plt.title('Solar System Data for Question 3')
plt.ylabel('Planetary Mass [Mjup]')
plt.xlabel('Semi-Major Axis')
plt.grid()
plt.legend()
plt.show()
```
